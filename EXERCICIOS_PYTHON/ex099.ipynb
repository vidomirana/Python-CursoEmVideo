```python
from time import sleep
lista_num = []
while True:
    n = int(input('Adicione um número (0 para): '))
    if n == 0:
        break
    lista_num.append(n)


def analise_nums(lista):
    lista.sort()
    print('-=' * 25)
    for v in lista:
        print(v, end=' ')
        sleep(0.4)
    print()
    print(f'Foram digitados {len(lista)} números')
    print(f'O maior número foi {max(lista)} e o menor {min(lista)}')


analise_nums(lista_num)
```
