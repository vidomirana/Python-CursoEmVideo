```python
from time import sleep
from random import randint
numeros = []


def sorteio():
    print('-=' * 30)
    for i in range(5):
        numeros.append(randint(1, 21))
    print(f'Foram sorteados os números ', end='')
    for n in numeros:
        print(n, end=' ', flush=True)
        sleep(0.5)
    print()


def somaPar(lista):
    sp = 0
    for v in lista:
        if v % 2 == 0:
            sp += v
    print(f'A soma dos pares é {sp}')


sorteio()
somaPar(numeros)
```
